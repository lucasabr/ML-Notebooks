#### Check for outliers if any and handle them

To check for outliers, we will graph a boxplot for every feature. the x axis will be our target variable 'quality', while the y axis will be the variable we are looking for outliers in. Boxplot uses IQR to check for outliers, and any dots outside of each box, are considered outliers.

The reason we are plotting these features along with the target variable, rather than by itself, it that the target variable is what we are trying to predict. 

For example, an entry with a pH of 2.8 may be an outlier when plotted individually, but when plotted with its quality of 3, it is actually not an outlier. On the contrary, an entry with a pH of 3.5 may not be an outlier when plotted individually, but when plotted with a quality of 4, it might be considered one.

```python
# Outliers detection and class imbalance
def boxplots():
    fig, axes = plt.subplots(nrows=12,ncols=1)
    fig.set_size_inches(20, 100)
    for i, feature in enumerate(wine.columns):
        sns.boxplot(data=wine,y=feature,x="quality",orient="v",ax=axes[i])
boxplots()
```

As shown above, there are quite a few outliers in our data. We will use the below code to find all these outliers (using IQR), and remove them. We will then do boxplots again to check the data. 

While there may still be some outliers in this new data, due to the dataset changing, most of them should be removed, and the ones that are there would be less extreme.

```python
def outliers_removal(feature):
    global wine
    
    for i in range(3,9):
        dataset = wine[wine['quality'] == i]
        data = dataset[feature]
        
        q25, q75 = np.percentile(data, 25), np.percentile(data, 75)
        iqr = q75 - q25
        
        cut_off = iqr * 1.5
        lower, upper = q25 - cut_off, q75 + cut_off
        outliers = [x for x in data if x < lower or x > upper]  
        wine.drop(wine[(wine[feature].isin(outliers)) & (wine['quality']==i)].index, inplace=True)
        
for feat in wine.columns:
    outliers_removal(feat)
boxplots()
```

### Modeling

Before we create any models, we need to split the data into training data, and test data. We will use 80% of the original data to train our models, and 20% to test our models.

```python
from sklearn.model_selection import train_test_split
features = wine.drop('quality', axis=1)
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.2, random_state=42)
```

#### Implement Multivariable Linear Regression using any 2 Regression algorithms of your choice

The 2 Linear Regression algorithms we will use is LassoCV, and RidgeCV. Lasso and Ridge are linear regression algorithms that include some form of regularization, to discourage and hopefully prevent our models from overfitting to the training data.

We will use LassoCV and RidgeCV rather than Lasso and Ridge, as these 2 algorithms have 5Fold Cross Validation built into them, making the training of our data even better.

```python
from sklearn.linear_model import LassoCV, RidgeCV

lasso = LassoCV()
lasso.fit(xtrain, ytrain)
results1 = lasso.predict(xtest)

ridge = RidgeCV()
ridge.fit(xtrain, ytrain)
results2 = ridge.predict(xtest)
```

#### Implement a Neural Network using Tensorflow and Keras to classify wine quality

We will implement a neural network with similar layers to those done in lectures.

```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

n_classes = 6
mapping = {3:0, 4:1, 5:2, 6:3, 7:4, 8:5}
y_train = ytrain.copy()
y_test = ytest.copy()
y_train = y_train.map(mapping)
y_test = y_test.map(mapping)
y_train = to_categorical(y_train, n_classes)
y_test = to_categorical(y_test, n_classes)

print(y_train[1])
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=11))
model.add(Dense(64, activation='relu'))
model.add(Dense(6, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer=SGD(learning_rate=0.1), metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
model.summary()
```

```python
history = model.fit(xtrain, y_train, batch_size=128, epochs=50, verbose=1, validation_data=(xtest, y_test))
```

### Model Evaluation

Evaluating the model accuracy is an essential part of the process in creating machine learning models to describe how well the model is performing in its predictions. Evaluation metrics change according to the problem type. Here, we'll briefly learn how to check the accuracy of the regression model.

The linear model (regression) can be a typical example of this type of problem, and the main characteristic of the regression problem is that the targets of a dataset contain the real numbers only. The errors represent how much the model is making mistakes in its prediction. The basic concept of accuracy evaluation is to compare the original target with the predicted one according to certain metrics.


**Regression model evaluation metrics**

The MAE, RMSE, and R-Squared metrics are mainly used to evaluate the prediction error rates and model performance in regression analysis.

**MAE (Mean absolute error)** represents the difference between the original and predicted values extracted by averaged the absolute difference over the data set.

**RMSE (Root Mean Squared Error)** is the error rate by the square root of MSE.

**R-squared (Coefficient of determination)** represents the coefficient of how well the values fit compared to the original values. The value from 0 to 1 interpreted as percentages. The higher the value is, the better the model is.

#### Compare the following 3 metrics: R-Squared, RMSE, and MAE between our two regression models and analyze the performance

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def print_metrics(name, prediction, result):
    print(name)
    print("\tMean Absolute Error - ", mean_absolute_error(result,prediction))
    print("\tRoot Mean Squared Error - ", mean_squared_error(result, prediction, squared = False))
    print("\tR Squared - ", r2_score(result, prediction))

print_metrics('LassoCV', results1, ytest)
print_metrics('RidgeCV', results2, ytest)
```

As you can see above, the 2 models perform very similarly, however the LassoCV model performs slightly better. The MEA and RMSE are slightly lower (by .002 and .003 respectively), and the R Squared is higher by .007 (for R Squared, the higher the value is the better the model is performing).

However, both these models perform extremely similarly, and the differences between them are extremely small.

#### Calculate the feature importance scores for the top features that help predicting wine quality and visualize them

```python
def plot_feature_importance(title, coefficients):
    plt.figure(figsize=(20,4))
    plt.bar(features.columns, coefficients)
    plt.title(title)
    plt.show()
plot_feature_importance('LassoCV', lasso.coef_)
plot_feature_importance('RidgeCV', ridge.coef_)
```

Shown above are the coefficients of the features in each of the regression models we have created.

On first glance, we can see that LassoCV has some form of feature selection, with 3 of the 10 variables having a coefficient of 0. On the other hand, RidgeCV has a coefficient set for every single feature, even if its an extremely small value (like that of residual sugar).

For both of these algorithms, the features with the highest positive coefficients are alcohol and sulphates, with Alcohol having the highest importance by a significant margin. This is interesting as it aligns with the EDA we performed earlier, where we examined the correlations in our data, and saw that sulphates and alcohol had the highest positive correlation with quality.

Another interesting similarity between these graphs and our correlation chart from before, is that in the correlation chart, volatile acidity had the highest negative correlation with quality, and here it has the highest negative coefficient. 

Overall, the feature importance we can see above it pretty similar and in line with the correlation analysis we did earlier.
However, it is not clear cut. For instance, citric acid had the 3rd highest positive correlation (0.23), and here it is a negative coefficient in one of our models.

Chlorides, total sulfur dioxide, and pH are the 3 features that both models have in common, and seem like minor, but still significant predictors.

#### Plot training loss and validation loss and Plot training accuracy and validation accuracy

```python
def plot_training_metrics(history, name, metric1, metric2):
    plt.plot(history.history[metric1])
    plt.plot(history.history[metric2])
    plt.title("Model " + name)
    plt.ylabel(name)
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.show()
plot_training_metrics(history, 'Accuracy', 'accuracy', 'val_accuracy')
plot_training_metrics(history, 'Loss', 'loss', 'val_loss')
```

Both the model accuracy and model loss graphs provide results that I would expect to see. 

In the accuracy graph, both train accuracy and validation accuracy start very low, but increase sharply. Train accuracy continues to slowly increase, while validation accuracy seems to plateau. Training accuracy (other than at the very beginning) is always somewhat higher than the validation accuracy, which makes sense as the model is fitting to the training data.

In the loss graph, the inverse is true. At first, train and validation loss are both very high, but decrease sharply. While training loss continues to decrease over time, the validation loss plateaus. In this case, training loss is always somewhat lower than validation accuracy, which makes sense because our model uses SGD, a version of gradient descent, which seeks to minimize loss for its training data.
